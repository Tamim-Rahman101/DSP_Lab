```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#Sine Wave
def sine_wave(amplitude, frequency, phase):
    phase_radian = phase * np.pi/180
    x = np.linspace(start=-2, stop=2, num=1000)
    y = amplitude * np.sin(2 * np.pi * frequency * x + phase_radian)
    plt.title(f'Sine Wave with A={amplitude}, F={frequency}, P={phase}')
    plt.xlabel("Time")
    plt.ylabel("Amplitude")
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.plot(x, y)
    plt.show()

sine_wave(4, 1, 0)
```

```python
#Cosine Wave
def cosine_wave(amplitude, frequency, phase):
    phase_radian = phase * np.pi/180
    x = np.linspace(start=-2, stop=2, num=1000)
    y = amplitude * np.cos(2 * np.pi * frequency * x + phase_radian)
    plt.title(f'Cosine Wave with A={amplitude}, F={frequency}, P={phase}')
    plt.xlabel("Time")
    plt.ylabel("Amplitude")
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.plot(x, y)
    plt.show()

cosine_wave(4, 1, 0)
```

```python
#Unit Sample / Impulse Signal
def unit_sample(start, end, sampling_rate):
    x = np.linspace(start=start,stop=end,num=int((end-start)*sampling_rate) + 1)
    y = []
    for i in x:
        if i==0:
            y.append(1)
        else:
            y.append(0)
    plt.title(f'Unit Sample Signal from {start} to {end}')
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.stem(x,y)
    plt.show()
unit_sample(-3, 3, 3)
```

```python
#Unit Step Signal
def unit_step(start, end, sampling_rate):
    x = np.linspace(start=start,stop=end,num=int((end-start)*sampling_rate) + 1)
    y = []
    for i in x:
        if i<0:
            y.append(0)
        else:
            y.append(1)
    plt.title(f'Unit Step Signal from {start} to {end}')
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.stem(x,y)
    plt.show()
unit_step(-3, 3, 3)
```

```python
#Unit Ramp Signal
def unit_ramp(start, end, sampling_rate):
    x = np.linspace(start=start,stop=end,num=int((end-start)*sampling_rate) + 1)
    y = []
    for i in x:
        if i<0:
            y.append(0)
        else:
            y.append(i)
    plt.title(f'Unit Ramp Signal from {start} to {end}')
    plt.grid(True, which='both')
    plt.axhline(color='r')
    plt.axvline(color='r')
    plt.stem(x,y)
    plt.show()
unit_ramp(-3, 3, 2)
```
