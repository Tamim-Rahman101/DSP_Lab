### Show that 50 Hz is an alias of the frequency 10 Hz, when sampling at 40 Hz. Assume the signals are
#### sin(2π.10t)
#### sin(2π.50t)

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
plt.figure(figsize=(10, 4))

t = np.linspace(start=0, stop=1, num=1000)
x_t = np.sin(2*np.pi*10*t)
plt.subplot(2, 2, 1)
plt.plot(t, x_t)
plt.grid(True, which='both')
plt.title('x(t) = sin(2π.10t)')
plt.xlabel('Time')
plt.ylabel('Amplitude')

t = np.linspace(start=0, stop=1, num=1000)
x_t = np.sin(2*np.pi*50*t)
plt.subplot(2, 2, 2)
plt.plot(t, x_t)
plt.grid(True, which='both')
plt.title('x(t) = sin(2π.50t)')
plt.xlabel('Time')
plt.ylabel('Amplitude')

n = np.arange(0, 1, 1/40)
x_n = np.sin(2*np.pi*10*n)
plt.subplot(2, 2, 3)
plt.stem(n, x_n)
plt.grid(True, which='both')
plt.title('x(n) = sin(2π.(10/40)n)')
plt.xlabel('Time')
plt.ylabel('Amplitude')

n = np.arange(0, 1, 1/40)
x_n = np.sin(2*np.pi*50*n)
plt.subplot(2, 2, 4)
plt.stem(n, x_n)
plt.grid(True, which='both')
plt.title('x(n) = sin(2π.(50/40)n)')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()

```
