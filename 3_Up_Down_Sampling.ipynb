```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(start=0, stop=0.008, num=10000)
x = 5*np.sin(2 * np.pi * 500 * t + np.radians(90))

plt.figure(figsize=(10, 6))

plt.subplot(3, 2, 1)
plt.plot(t, x)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('x(t) = 5sin(2pi.500t + pi/2)')
plt.grid(True, which='both')

fs = 3500
n = np.arange(0, 0.008, 1/fs)
x_n = 5*np.sin(2 * np.pi * 500 * n + np.radians(90))
plt.subplot(3, 2, 2)
plt.stem(n, x_n)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('x(n) = 5sin(2pi.500n + pi/2)')
plt.grid(True, which='both')

I = 2
up_sampled_x = np.zeros(I * len(x_n))
up_sampled_x[::2] = x_n
plt.subplot(3, 2, 3)
plt.stem(up_sampled_x)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Up-Sampled Signal')
plt.grid(True, which='both')

D = 2
down_sampled_x = x_n[::2]
plt.subplot(3, 2, 4)
plt.stem(down_sampled_x)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Down-Sampled Signal')
plt.grid(True, which='both')

# Smoothing the signal by taking the average of two adjacents of zero
for i in range(1, len(up_sampled_x)-1, 2):
    up_sampled_x[i] = (up_sampled_x[i-1] + up_sampled_x[i+1])/2
plt.subplot(3, 2, 5)
plt.stem(up_sampled_x)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Up-Sampled Smooth Signal')
plt.grid(True, which='both')

plt.tight_layout()
plt.show()
```
