```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convolve(x, h):
    len_x = len(x)
    len_h = len(h)
    len_y = len_x + len_h - 1
    y = [0] * len_y
    for n in range(len_y):
        for k in range(len_x):
            if(n - k >= 0 and n - k < len_h):
                y[n] += x[k] * h[n - k]
    return y
```

```python
# x = [1, 2, 3, 4, 5]
# h = [2, 3, 4]
x = [1, 2, 1, -1]
h = [1, 2, 3, 1]
h_folded = h[::-1]
convolution = convolve(x, h)
correlation = convolve(x, h_folded)

plt.figure(figsize=(10, 4))

plt.subplot(2, 3, 1)
plt.stem(range(len(x)), x)
plt.title('x(n)')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True, which='both')

plt.subplot(2, 3, 2)
plt.stem(range(len(h)), h)
plt.title('h(n)')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True, which='both')

plt.subplot(2, 3, 3)
plt.stem(range(len(convolution)), convolution)
plt.title('y(n)=x(n)*h(n)')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True, which='both')

plt.subplot(2, 3, 4)
plt.stem(range(len(x)), x)
plt.title('x(n)')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True, which='both')

plt.subplot(2, 3, 5)
plt.stem(range(len(h_folded)), h_folded)
plt.title('h(-n)')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True, which='both')

plt.subplot(2, 3, 6)
plt.stem(range(len(correlation)), correlation)
plt.title('y(n) = x(n)*h(-n)')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True, which='both')

plt.tight_layout()
plt.show()
```
